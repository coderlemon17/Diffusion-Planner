```python
import os
import hydra

os.environ["NUPLAN_DEVKIT_ROOT"] = "/home/zhouxinning/Workspace/Diffusion-Planner/nuplan-devkit"
os.environ["NUPLAN_DATA_ROOT"] = "/data/zhouxinning/nuPlan/nuplan/dataset"
os.environ["NUPLAN_MAPS_ROOT"] = "/data/zhouxinning/nuPlan/nuplan/dataset/maps"
os.environ["NUPLAN_EXP_ROOT"] = "/data/zhouxinning/nuPlan/nuplan/exp"
os.environ["NUPLAN_TUTORIAL_PATH"] = "../nuplan-devkit/tutorials"
```

```python
from tutorials.utils.tutorial_utils import construct_nuboard_hydra_paths

BASE_CONFIG_PATH = os.path.join(os.getenv('NUPLAN_TUTORIAL_PATH', ''), '../nuplan/planning/script')
# Location of paths with all nuBoard configs
nuboard_hydra_paths = construct_nuboard_hydra_paths(BASE_CONFIG_PATH)

# Initialize configuration management system
hydra.core.global_hydra.GlobalHydra.instance().clear()  # reinitialize hydra if already initialized
hydra.initialize(config_path=nuboard_hydra_paths.config_path)

# Compose the configuration
cfg = hydra.compose(config_name=nuboard_hydra_paths.config_name, overrides=[
    'scenario_builder=nuplan_mini',  # set the database (same as simulation) used to fetch data for visualization
    f'simulation_path={["/data/zhouxinning/nuPlan/nuplan/exp/exp/simulation/closed_loop_reactive_agents/diffusion_planner/val14/diffusion_planner_release/model_2025-02-17-11-00-46/nuboard_1739761275.nuboard"]}',  # nuboard file path, if left empty the user can open the file inside nuBoard
    f'hydra.searchpath=[{nuboard_hydra_paths.common_dir}, {nuboard_hydra_paths.experiment_dir}]',
    'port_number=5007'
])
```

```python
from nuplan.planning.script.run_nuboard import main as main_nuboard

# Run nuBoard
main_nuboard(cfg)
```
